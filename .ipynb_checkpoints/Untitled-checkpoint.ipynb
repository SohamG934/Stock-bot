```python
import streamlit as st
import pandas_datareader.data as web
import matplotlib.pyplot as plt
from io import BytesIO
import base64


class StockGrapher:
    def __init__(self, ticker):
        self.ticker = ticker
        self.data = web.DataReader(ticker, "yahoo")

    def plot1(self):
        plt.plot(self.data["Close"])
        plt.title(f"{self.ticker} Closing Prices")
        plt.xlabel("Date")
        plt.ylabel("Price ($)")
        buffer = BytesIO()
        plt.savefig(buffer, format="png")
        return base64.b64encode(buffer.getvalue()).decode()

    def plot2(self):
        plt.plot(self.data["Volume"])
        plt.title(f"{self.ticker} Trading Volume")
        plt.xlabel("Date")
        plt.ylabel("Volume")
        buffer = BytesIO()
        plt.savefig(buffer, format="png")
        return base64.b64encode(buffer.getvalue()).decode()

    def plot3(self):
        plt.plot(self.data["Open"], label="Open")
        plt.plot(self.data["Close"], label="Close")
        plt.plot(self.data["High"], label="High")
        plt.plot(self.data["Low"], label="Low")
        plt.title(f"{self.ticker} OHLC Prices")
        plt.xlabel("Date")
        plt.ylabel("Price ($)")
        plt.legend()
        buffer = BytesIO()
        plt.savefig(buffer, format="png")
        return base64.b64encode(buffer.getvalue()).decode()


def main():
    st.title("Stock Grapher")

    # User input for stock ticker
    ticker = st.text_input("Enter a stock ticker:")

    # Generate plots if a valid ticker is entered
    if ticker:
        try:
            grapher = StockGrapher(ticker)
            plot1_data = grapher.plot1()
            plot2_data = grapher.plot2()
            plot3_data = grapher.plot3()

            # Display plots
            st.image(plot1_data, use_column_width=True)
            st.image(plot2_data, use_column_width=True)
            st.image(plot3_data, use_column_width=True)

        except:
            st.error("Error: Please enter a valid stock ticker.")


if __name__ == "__main__":
    main()
import streamlit as st
import pandas_datareader.data as web
import matplotlib.pyplot as plt
from io import BytesIO
import base64


class StockGrapher:
    def __init__(self, ticker):
        self.ticker = ticker
        self.data = web.DataReader(ticker, "yahoo")

    def plot1(self):
        plt.plot(self.data["Close"])
        plt.title(f"{self.ticker} Closing Prices")
        plt.xlabel("Date")
        plt.ylabel("Price ($)")
        buffer = BytesIO()
        plt.savefig(buffer, format="png")
        return base64.b64encode(buffer.getvalue()).decode()

    def plot2(self):
        plt.plot(self.data["Volume"])
        plt.title(f"{self.ticker} Trading Volume")
        plt.xlabel("Date")
        plt.ylabel("Volume")
        buffer = BytesIO()
        plt.savefig(buffer, format="png")
        return base64.b64encode(buffer.getvalue()).decode()

    def plot3(self):
        plt.plot(self.data["Open"], label="Open")
        plt.plot(self.data["Close"], label="Close")
        plt.plot(self.data["High"], label="High")
        plt.plot(self.data["Low"], label="Low")
        plt.title(f"{self.ticker} OHLC Prices")
        plt.xlabel("Date")
        plt.ylabel("Price ($)")
        plt.legend()
        buffer = BytesIO()
        plt.savefig(buffer, format="png")
        return base64.b64encode(buffer.getvalue()).decode()


def main():
    st.title("Stock Grapher")

    # User input for stock ticker
    ticker = st.text_input("Enter a stock ticker:")

    # Generate plots if a valid ticker is entered
    if ticker:
        try:
            grapher = StockGrapher(ticker)
            plot1_data = grapher.plot1()
            plot2_data = grapher.plot2()
            plot3_data = grapher.plot3()

            # Display plots
            st.image(plot1_data, use_column_width=True)
            st.image(plot2_data, use_column_width=True)
            st.image(plot3_data, use_column_width=True)

        except:
            st.error("Error: Please enter a valid stock ticker.")


if __name__ == "__main__":
    main()
```
